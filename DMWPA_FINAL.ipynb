# Q2
## By how much do the ESG funds reduce carbon emissions compared to other funds/indices?

### Fund sectors contribution




```python
# Filter data
sector_data = fact_df.loc[3:, ['X', 'S.P.500', 'Dow.Jones.U.S..Large.Cap.Total.Stock.Market.Index',
                            'Narcissus.Core.Equity.Sustainability.Fund',
                            'Pietro.Advisory.Sustainable.Large.Cap.ETF']]
sector_data.columns = ['Sector', 'S&P 500', 'Dow Jones', 'Narcissus Core Equity Sustainability Fund', 'Pietro Advisory Sustainable Large Cap ETF']
sector_data.dropna(inplace=True)

sectors = sector_data['Sector'].tolist()

# Dictionary to hold data for each index for looping
indices = {
    'S&P 500': sector_data['S&P 500'].tolist(),
    'Dow Jones': sector_data['Dow Jones'].tolist(),
    'Narcissus Core Equity Sustainability Fund': sector_data['Narcissus Core Equity Sustainability Fund'].tolist(),
    'Pietro Advisory Sustainable Large Cap ETF': sector_data['Pietro Advisory Sustainable Large Cap ETF'].tolist(),
}

for index_name, values in indices.items():
    total = sum(values)
    values.append(total)

    x_labels = sectors + ["Total"]

    measures = ["relative"] * (len(values) - 1) + ["total"]

    text = []
    for v in values:
        text.append(str(v))

    # Create the waterfall chart
    fig = go.Figure(go.Waterfall(
        name=f"{index_name} Sector Contributions", orientation="v",
        measure=measures,
        x=x_labels,
        textposition="outside",
        text=text,
        y=values,
        connector={"line": {"color": "rgb(63, 63, 63)"}}
    ))

    fig.update_layout(
        title=f"{index_name} Sector Contributions",
        showlegend=True
    )

    # Display the chart
    fig.show()
```

The graph shown in the figure above is called a waterfall chart, the purpose of the waterfall chart within context of our poster is to visualize the contributions of each sector within the indices. The reason why we wanted to use waterfall chart instead of other charts is because we want to let the viewers of our poster understand the weight of each sector within the indices in a more dynamic way.

To create this chart, we had to look up waterfall chart documentation in the plotly library, which is the library used to create this chart. Besides that, we also had to watch youtube videos covering the waterfall chart on youtube

### Overall Reduction of Emmisions


```python
# Define the ESG and non-ESG funds/indices
ESG = ['Narcissus.Core.Equity.Sustainability.Fund', 'Pietro.Advisory.Sustainable.Large.Cap.ETF']
non_ESG = ['S.P.500', 'Dow.Jones.U.S..Large.Cap.Total.Stock.Market.Index']
esg_emissions = fact_df.loc[0:1, ESG]
nonesg_emissions = fact_df.loc[0:1, non_ESG]

# Calculate the average emissions for ESG and non-ESG across different metrics
avg_esg_emissions = esg_emissions.mean().mean()
avg_nonesg_emissions = nonesg_emissions.mean().mean()

# Calculate the percentage reduction in emissions for ESG funds
reduction_percentage = ((avg_nonesg_emissions - avg_esg_emissions) / avg_nonesg_emissions) * 100
```

```python
metrics = ['Fossil Fuel Emissions', 'Carbon to Value Invested']
esg_values = esg_emissions.mean(axis=1).values
nonesg_values = nonesg_emissions.mean(axis=1).values

# Create a stacked bar chart
fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.4
index = np.arange(len(metrics))

# Non-ESG Emissions as the full bars
ax.bar(index, nonesg_values, bar_width, label='Non-ESG Total Emissions', color='red', alpha=0.7)
ax.bar(index, esg_values, bar_width, label='ESG Emissions', color='green')
ax.set_xlabel('Emission Metric')
ax.set_ylabel('Carbon Emissions (Metric Tons per $1M invested)')
ax.set_title('Carbon Emissions by Metric: ESG vs Non-ESG Funds')
ax.set_xticks(index)
ax.set_xticklabels(metrics)
ax.legend()

plt.tight_layout()
plt.show()
```

```python
# Gauge chart to visualize the reduction percentage
fig_gauge = go.Figure(go.Indicator(
    mode="gauge+number",
    value=reduction_percentage,
    title={'text': "Carbon Emissions by Metric difference (%)"},
    gauge={
        'axis': {'range': [0, 100], 'tickwidth': 1, 'tickcolor': "darkblue"},
        'bar': {'color': "purple"},
        'steps': [
            {'range': [0, 50], 'color': "lightgray"},
            {'range': [50, 100], 'color': "lightgreen"}
        ],
        'threshold': {
            'line': {'color': "red", 'width': 4},
            'thickness': 0.75,
            'value': reduction_percentage}
    }
))


# Show the figures
fig_gauge.show()
```

### Yearly Percentage Change in Closing Price for Indices


```python
pfc_df_mdf = pfc_df.drop(columns=['S.P.500.Energy..Sector.','S.P.500.Utilities..Sector.','S.P.500.Communication.Services..Sector.','S.P.500.Consumer.Discretionary..Sector.','S.P.500.Consumer.Staples..Sector.','S.P.500.Financials..Sector.','S.P.500.Health.Care..Sector.','S.P.500.Industrials..Sector.','S.P.500.Information.Technology..Sector.','S.P.500.Materials..Sector.','S.P.500.Real.Estate..Sector.'])
pfc_df_mdf=pfc_df.drop(columns=['Effective.date.1'])
df =pfc_df_mdf.loc[[0,4,5,16,17,28,29,40,41,52,53,64,65,76,77,88,89,100,101,112,113,120]]
```

```python
# Convert the 'Effective.date' column to datetime
df['Effective.date'] = pd.to_datetime(df['Effective.date'])
df['Year'] = df['Effective.date'].dt.year

# Calculate the percentage change within each year for each column of interest
difference = ['S.P.500', 'Narcissus.Core.Equity.Sustainability.Fund..NAV.in...',
                        'Pietro.Advisory.Sustainable.Large.Cap.ETF..NAV.in...',
                        'Dow.Jones.U.S..Large.Cap.Total.Stock.Market.Index']

# Calculate percentage change within each year for the selected columns
df_pct_change = df.groupby('Year')[difference].pct_change() * 100

for col in difference:
    df[f'{col}_Pct_Change'] = df_pct_change[col]
```

```python
columns_to_keep = ['S.P.500_Pct_Change','Narcissus.Core.Equity.Sustainability.Fund..NAV.in..._Pct_Change','Pietro.Advisory.Sustainable.Large.Cap.ETF..NAV.in..._Pct_Change','Dow.Jones.U.S..Large.Cap.Total.Stock.Market.Index_Pct_Change']
df= df[columns_to_keep].dropna()
years = ['2014','2015','2016','2017','2018','2019','2020','2021','2022','2023','2024']
df['year']= years
```

```python
# Plotting each column against the 'year' column
plt.figure(figsize=(12, 8))

plt.plot(df['year'], df['S.P.500_Pct_Change'], label='S&P 500 % Change')
plt.plot(df['year'], df['Narcissus.Core.Equity.Sustainability.Fund..NAV.in..._Pct_Change'], label='Narcissus Core Equity Fund % Change')
plt.plot(df['year'], df['Pietro.Advisory.Sustainable.Large.Cap.ETF..NAV.in..._Pct_Change'], label='Pietro Advisory Sustainable ETF % Change')
plt.plot(df['year'], df['Dow.Jones.U.S..Large.Cap.Total.Stock.Market.Index_Pct_Change'], label='Dow Jones US Large Cap % Change')
plt.xlabel("Year")
plt.ylabel("Percentage Change (%)")
plt.title("Yearly Percentage Change Comparison of Various Funds and Indices")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Setting up the DataFrame for bar plotting
df.set_index('year', inplace=True)

# Plotting the bar chart
df[['S.P.500_Pct_Change',
    'Narcissus.Core.Equity.Sustainability.Fund..NAV.in..._Pct_Change',
    'Pietro.Advisory.Sustainable.Large.Cap.ETF..NAV.in..._Pct_Change',
    'Dow.Jones.U.S..Large.Cap.Total.Stock.Market.Index_Pct_Change']].plot(kind='bar', figsize=(14, 8))
plt.xlabel("Year")
plt.ylabel("Percentage Change (%)")
plt.title("Yearly Percentage Change Comparison of Various Funds and Indices")
plt.legend(["S&P 500 % Change",
            "Narcissus Core Equity Fund % Change",
            "Pietro Advisory Sustainable ETF % Change",
            "Dow Jones US Large Cap % Change"])
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

### Normalized Price Trends of Portfolios

```python
new_df = pfc_df.drop(columns=['S.P.500.Energy..Sector.','S.P.500.Utilities..Sector.','S.P.500.Communication.Services..Sector.','S.P.500.Consumer.Discretionary..Sector.','S.P.500.Consumer.Staples..Sector.','S.P.500.Financials..Sector.','S.P.500.Health.Care..Sector.','S.P.500.Industrials..Sector.','S.P.500.Information.Technology..Sector.','S.P.500.Materials..Sector.','S.P.500.Real.Estate..Sector.'])

# Select only the numeric columns
numeric_df = new_df.select_dtypes(include='number')

# Normalize each portfolio's price to start at 100
df_normalized = numeric_df.divide(numeric_df.iloc[0]) * 100

# Plot the normalized data
df_normalized.plot(figsize=(12, 8))
plt.xlabel("Date")
plt.ylabel("Normalized Price (Base 100)")
plt.title("Normalized Price Trends of Portfolios")
plt.legend(title="Portfolios")
plt.show()
```

```python
# Min-max scale each portfolio's price to a range of 0 to 1
pfc_df_min_max_scaled = (numeric_df - numeric_df.min()) / (numeric_df.max() - numeric_df.min())

# Plot the min-max scaled data
pfc_df_min_max_scaled.plot(figsize=(12, 8))
plt.xlabel("Date")
plt.ylabel("Scaled Price (0 to 1)")
plt.title("Min-Max Scaled Price Trends of Portfolios")
plt.legend(title="Portfolios")
plt.show()
```
